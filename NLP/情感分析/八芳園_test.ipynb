```python
import pandas as pd
df = pd.read_csv("E:/專題/景點爬蟲/tripadvisor/tripadvisor/情感分析/tripadvisor/八芳園",encoding="utf-8")
df
```

```python
content = df["評論"]  
content
```

```python
import re
def del_word(x):
    s = (x.replace("透過行動裝置","").replace("[","").replace("]","").replace(",","").replace("...更多","")
          .replace("'","").strip()).replace(' 由 Asia Online Language Studio 翻譯 評等翻譯',"").replace("TripAdvisor 會員","")

    result = re.split('體驗日期\W\s\d+年\d+月',s)
    result2 = re.split('感謝\s+\w+', "".join(result)) 
    
    return result2

comment = []
for i in content:
    b = del_word(''.join(i))
    comment.append(b)

    
comment
```

```python
df["comment"] = comment 
df = df.drop(columns=["評論"])
```

```python
df
```

```python
star = df["評論星等"] 
stars = []
for i in star:
#     print(i)
    a = i.replace("ui_bubble_rating bubble_", "")
    stars.append(a)
```

```python
df["Star"] = stars
df = df.drop(columns=["評論星等"])
df
```

```python
pos=''
neg=''
for i in range(df.shape[0]):
    if df.loc[i,"Star"] in ["40","50"]:
        pos+=(df.loc[i,'comment'][0]+'\n')
    else:
        neg+=(df.loc[i,'comment'][0]+'\n')
```
