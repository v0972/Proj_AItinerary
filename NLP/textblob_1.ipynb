```python
content = df["Comment"]
content
```

```python
import re

def del_word(x):
  s = (x.replace("ÈÄèÈÅéË°åÂãïË£ùÁΩÆ", "").replace("[", "").replace("]", "").replace(",", "").replace("...More", "")
       .replace("'", "").strip()).replace("TripAdvisor ÊúÉÂì°", "")

  result = re.split('Date\s+of\s+experience\W\s\w+\s\d+', s)
  result2 = re.split('Thank\s+\w+', "".join(result))
  return result2
```

```python
comment = []
for i in content:
    b = del_word(i)
    comment.append(b)
    
df = df.drop(columns = ["Comment"])
df["Comment"] = comment


```

```python
star = df["Comment_Star"]
stars = []
for z in star:
    a = z.replace("ui_bubble_rating bubble_", "")
    stars.append(a)

df["Star"] = stars
df = df.drop(columns=["Comment_Star"])
```
